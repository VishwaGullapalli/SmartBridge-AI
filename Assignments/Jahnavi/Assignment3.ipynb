```python
from tensorflow.keras.layers import Dense,Flatten,Input
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
```

```python
train_data = '/kaggle/input/bird-species-classification/train_data/train_data'
test_data ='/kaggle/input/bird-species-classification/test_data/test_data'
```

```python
train_gen = ImageDataGenerator(rescale =1./225,
                               shear_range=0.2,
                               zoom_range=0.2,
                               horizontal_flip=True)

test_gen = ImageDataGenerator(rescale =1./225)
```

```python
train = train_gen.flow_from_directory(train_data,
                                      target_size=(224, 224),
                                      batch_size=22,
                                      class_mode='categorical')

test = test_gen.flow_from_directory(test_data,
                                    target_size=(224, 224),
                                    batch_size=22,
                                    class_mode='categorical')
```

```python
from tensorflow.keras.applications.vgg16 import VGG16,preprocess_input
```

```python
vgg = VGG16(include_top=False,
            weights='imagenet', 
            input_shape=(224,224,3))
```

```python
for layer in vgg.layers:
  print(layer)
```

```python
for layer in vgg.layers:
  layer.trainable=False
```

```python
x = Flatten()(vgg.output)
```

```python
prediction = Dense(16,activation='softmax')(x)
```

```python
model = Model(inputs=vgg.input,
              outputs=prediction)
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit_generator(train,validation_data=test,epochs=10,steps_per_epoch=len(train),validation_steps=len(test))
```
